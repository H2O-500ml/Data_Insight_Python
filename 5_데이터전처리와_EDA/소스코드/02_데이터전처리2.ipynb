'Target'에 관련한 내용들을 좀 더 실습 합시다.

**다음 내용을 데이터 프레임에 반영하거나, 따로 선언해두십시요.**

2. 레몬과 오렌지의 각각의 원가는 0.05입니다.
3. 레몬은 메일 200개, 오렌지는 150개씩 구매하여 사용합니다. 하루가 지나면 폐기됩니다.
    * lemon_prepared, orange_prepared 라는 컬럼을 만들어 추가하세요.
    * lemon_total_cost, orange_total_cost라는 컬럼을 만들어 원가비용을 추가하세요.
4. 일별 이익을 계산한 컬럼이 필요합니다.
    * net_profit 라는 컬럼을 만들어 추가하세요.
5. 폐기하는 레몬과 오렌지 개수를 계산한 컬럼이 필요합니다.
    * x_orange, x_lemon 이라는 컬럼을 만들어 추가하세요.
6. 레몬과 오렌지를 버리기 때문에 생기는 손실액을 계산하는 컬럼이 필요합니다.(폐기 손실)
    * 예를 들어 준비한 레몬 150개 중, 120개를 팔았다면, 30개는 버려야 합니다. 30개 x 0.05 = 1.5 의 손해가 발생했다고 이야기 합시다.
    * loss_orange, loss_lemon 이라는 컬럼을 만들어 추가하세요.
    * 총 손실에 대해서 loss_total이라는 컬럼을 만들어 추가하세요.
    
    
**다시 한 번, Optimize의 Target이 무엇인가요?**
**easy> 무엇을 높여야 하고, 무엇을 줄여야 하나요?**


````python
# 아래 사항에 맞게 변수를 담습니다.
# 1. 하루 총 운영비용이 5 가 소요됩니다.  
# 2. 레몬과 오렌지의 각각의 원가는 0.05입니다.

temp1['operation_cost'] = 5
temp1['unit_cost'] = 0.05
````

````python
# 3. 레몬은 메일 200개, 오렌지는 150개씩 구매하여 사용합니다. 하루가 지나면 폐기됩니다.
#     * lemon_prepared, orange_prepared 라는 컬럼을 만들어 추가하세요.
#     * lemon_total_cost, orange_total_cost라는 컬럼을 만들어 원가비용을 추가하세요.

temp1['lemon_prepared'] = 200
temp1['orange_prepared'] = 150
temp1['lemon_total_cost'] = temp1['lemon_prepared']*temp1['unit_cost']
temp1['orange_total_cost'] = temp1['orange_prepared']*temp1['unit_cost']
temp1.head()
````

````python
# 4. 일별 이익을 계산한 컬럼이 필요합니다.
#     * net_profit 라는 컬럼을 만들어 추가하세요.

temp1['net_profit']=temp1['total_amt']-(temp1['lemon_total_cost']+temp1['orange_total_cost'])

````

````python
# 5. 폐기하는 레몬과 오렌지 개수를 계산한 컬럼이 필요합니다.
#     * x_orange, x_lemon 이라는 컬럼을 만들어 추가하세요.

temp1['x_lemon'] = temp1['lemon_prepared'] - temp1['Lemon']
temp1['x_orange'] = temp1['orange_prepared'] - temp1['Orange']
````

````python
# 6. 레몬과 오렌지를 버리기 때문에 생기는 손실액을 계산하는 컬럼이 필요합니다.(폐기 손실)
#     * 예를 들어 준비한 레몬 150개 중, 120개를 팔았다면, 30개는 버려야 합니다. 30개 x 0.05 = 1.5 의 손해가 발생했다고 이야기 합시다.
#     * loss_orange, loss_lemon 이라는 컬럼을 만들어 추가하세요.
#     * 총 손실에 대해서 loss_total이라는 컬럼을 만들어 추가하세요.

temp1['loss_lemon'] = temp1['x_lemon']*temp1['unit_cost']
temp1['loss_orange'] = temp1['x_orange']*temp1['unit_cost']
temp1['loss_total'] = temp1['loss_lemon']+temp1['loss_orange']
````
