## 2. Customization

### Axis Label, Title

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

plt.plot(year, pop)

# label
plt.xlabel('Year')
plt.ylabel('Population')

# Title
plt.title('World Population Projections')

plt.show()
```

### Ticks

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

plt.plot(year, pop)

# label
plt.xlabel('Year')
plt.ylabel('Population')

# Title
plt.title('World Population Projections')

# Ticks
plt.yticks([0, 2, 4, 6, 8, 10]
           , ['0', '2B', '4B', '6B', '8B', '10B'])

plt.show()
```

### Color

```python
values = [0,1.6,1.4,2.2,2.5,2.6,3.2,3.5,3,3.9,4.2,6]
plt.hist(values, bins = 6, color="yellow")
plt.show()
```

### Common Axis

```python
year = [1950, 1970, 1990, 2010]
pop1 = [2.519, 3.692, 5.263, 6.972]
pop2 = [1.231, 2.252, 2.988, 4.334]

plt.plot(year, pop1, 'red')
plt.plot(year, pop2, 'blue')

# label
plt.xlabel('Year')
plt.ylabel('Population')

# Title
plt.title('World Population Projections')

plt.show()

```

### Subplot

```python
year = [1950, 1970, 1990, 2010]
pop1 = [2.519, 3.692, 5.263, 6.972]
pop2 = [1.231, 2.252, 2.988, 4.334]

# plot1
plt.subplot(2,1,1) # nrows, ncolumns, index - plot을 두 줄로 배열할 건 데 그 중에 첫 번째라는 뜻
plt.plot(year, pop1, 'red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Pop1')

# plot2
plt.subplot(2, 1, 2)
plt.plot(year, pop2, 'blue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Pop2')

plt.show()
```

```python
year = [1950, 1970, 1990, 2010]
pop1 = [2.519, 3.692, 5.263, 6.972]
pop2 = [1.231, 2.252, 2.988, 4.334]

# plot1
plt.subplot(2, 1, 1)
plt.plot(year, pop1, 'red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Pop1')

# plot2
plt.subplot(2, 1, 2)
plt.plot(year, pop2, 'blue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Pop2')

plt.tight_layout() # Or equivalently,  "plt.tight_layout()" - 위아래 간격 벌려줌

plt.show()
```

## 4. seaborn 패키지

seaborn 패키지는 차트를 더 다양하고 효과적으로 시각화하도록 돕는 패키지 입니다.


```python
# 패키지 로딩
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 4.1 distplot (Density & Histogram)

```python
# 데이터 불러오기
aq = pd.read_csv(r"C:\Users\50008313\Python_Jupyter_Study\2기교육\5_데이터전처리와_EDA\샘플데이터\airquality.csv", engine = 'python')
ti =  pd.read_csv(r"C:\Users\50008313\Python_Jupyter_Study\2기교육\5_데이터전처리와_EDA\샘플데이터\titanic_simple.csv", engine = 'python') 
```

```python
# Density plot만 그리기 위한 .kdeplot()
sns.kdeplot(aq["Ozone"])
plt.show()
```
