### SQL Server에서 가져온 Sales, Customers, Products 를 이용하여 실습을 진행하세요.

```python
import pandas as pd
```

```python
# 필요한 라이브러리를 로딩하시오
# Sales, Customers, Products.csv를 파일에서 읽어 오세요.

sales = pd.read_csv(r"C:\Users\50008313\Python_Jupyter_Study\2기교육\3_Python기초\샘플데이터\Sales.csv", engine = 'python')
customers = pd.read_csv(r"C:\Users\50008313\Python_Jupyter_Study\2기교육\3_Python기초\샘플데이터\Customers.csv", engine = 'python')
products = pd.read_csv(r"C:\Users\50008313\Python_Jupyter_Study\2기교육\3_Python기초\샘플데이터\Products.csv", engine = 'python')

sales['OrderDate'] = pd.to_datetime(sales['OrderDate'])
sales['Year'] = sales['OrderDate'].dt.year
sales['Month'] = sales['OrderDate'].dt.month
```

```python
#1) Sales 에서 연도별 매출액을 조회하시오.
sales.groupby('Year')['Amt'].sum()
```

```python
#2) 2016년 월별 매출액을 조회하시오.
sales[sales['Year'] == 2016].groupby('Month')['Amt'].sum()
```

```python
#3) 2017년 고객별 매출액 상위 10명을 조회하시오.
Customer_Amt = sales.groupby('CustomerID')[['Amt']].sum().reset_index()
```

```python
Customer_Amt.sort_values(by=['Amt'], ascending=False).head(10)
```

```python
#4) 2017년에 가장 많이 팔린 상품 top 5를 판매량과 함께 조회하시오.
ProductID_Qty = sales.groupby('ProductID')[['Qty']].sum().reset_index()
ProductID_Qty.sort_values(by=['Qty'], ascending=False).head(10)
```

```python
#5) 2016년 Customers의 Addr1(시도별) 별 매출액을 집계하시오.
pd.merge(sales, customers).head()
```

```python
pd.merge(sales, customers).groupby('Addr1')[['Amt']].sum().reset_index()
```

```python
# 6) 연도별 Products의 Category 별 매출액을 집계하여 저장하시오.

pd.merge(sales, products).head()
```

```python
pd.merge(sales, products).groupby(['Year','Category'])[['Amt']].sum()
```

```python
Amt_by_Year_Cat =  pd.merge(sales, products).groupby(['Year','Category'])[['Amt']].sum().reset_index()
Amt_by_Year_Cat
```

```python
# 7) Category는 범주형 데이터입니다. 
# 가변수(Dummy variable)화 하여 6번 결과와 합치고,
# 가변수 칼럼의 이름을 변경하시오.
df_cat = pd.get_dummies(Amt_by_Year_Cat['Category'])

Amt_by_Year_Cat2 = pd.concat([Amt_by_Year_Cat, df_cat], axis=1)

Amt_by_Year_Cat2.columns.values[3:] = ['Cat_간식', 'Cat_과일', 'Cat_반찬류', 'Cat_유제품','Cat_채소']

Amt_by_Year_Cat2.drop(['Category'], axis=1, inplace=True) 
Amt_by_Year_Cat2
```
